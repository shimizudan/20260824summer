## 樹形図の代わりに全部生成

Rの **expand.grid** は「全部の組合せの表」を一発で作ります。

```r
# 1,2,3で3桁:全27通りの表
g <- expand.grid(a = 1:3, b = 1:3, c = 1:3)
head(g, 10)
```

```r
nrow(g)  # 3^3 = 27通り
```

```r
# 文字列として表示
apply(g, 1, paste0, collapse = "")
```

## 順列:5人から3人選んで並べる

```r
people <- c("A", "B", "C", "D", "E")
p <- permutations(5, 3, people)
head(p, 10)
```

```r
nrow(p)  # ₅P₃ = 5×4×3 = 60
```

```r
c(factorial(5), nrow(permutations(5, 5, people)))  # 全員並べると5!
```

## 組合せ:選ぶだけなら?

組合せはR標準の **combn** で作れます。

```r
combn(people, 3)  # 各列が1つの選び方
```

```r
c(ncol(combn(people, 3)), choose(5, 3))  # ₅C₃ = 10
```

```r
# PとCの関係:60 ÷ 3! = 10
nrow(p) / factorial(3) == choose(5, 3)
```

## 重複順列・円順列

```r
# ○×クイズ5問:2^5 = 32通り
nrow(expand.grid(rep(list(c("○", "×")), 5)))
```

```r
# 円順列:Aを固定して残り4人を並べる → (5-1)!
q <- permutations(4, 4, c("B", "C", "D", "E"))
for (i in 1:nrow(q)) cat("A-", paste(q[i, ], collapse = "-"), "\n", sep = "")
factorial(4)
```

## 同じものを含む順列

```r
# A,A,B,B,B の並べ替え(set=FALSEで同じ文字を区別せず生成→uniqueで重複除去)
r <- permutations(5, 5, c("A","A","B","B","B"), set = FALSE)
u <- unique(apply(r, 1, paste0, collapse = ""))
print(u)
length(u)
```

```r
# 公式 5!/(2!3!) と比較
factorial(5) / (factorial(2) * factorial(3))
```

## パスカルの三角形

```r
for (n in 0:10) {
    row <- choose(n, 0:n)
    cat(strrep(" ", 2*(10 - n)), sprintf("%4d", row), "\n")
}
```

```r
# 各行の和は2^n になっている!
for (n in 0:10) cat(sprintf("n=%d: 和 = %d\n", n, sum(choose(n, 0:n))))
```

## 二項係数に色を塗ると…(シェルピンスキーのガスケット)

※ 大きなnではchoose()の誤差が出るため、ここではN=32で実験します。

```r
N <- 32
m <- outer(0:(N-1), 0:(N-1),
           function(n, r) ifelse(r <= n, choose(n, r) %% 2, 0))
image(t(m[N:1, ]), col = c("white", "blue"), axes = FALSE,
      main = "Pascal's triangle mod 2")
```
