## 計算チェック!

展開の公式:$(2\sqrt{3}+5)(\sqrt{3}-1) = 1 + 3\sqrt{3}$

```r
# 左辺と右辺
c((2*sqrt(3) + 5) * (sqrt(3) - 1), 1 + 3*sqrt(3))
```

```r
# ほぼ等しいかチェック(Juliaの≈に相当)
isTRUE(all.equal((2*sqrt(3) + 5) * (sqrt(3) - 1), 1 + 3*sqrt(3)))
```

## for文:繰り返しの魔法

```r
for (i in 1:5) {
    print(i)
}
```

```r
# 7の段の九九(catとsprintfで埋め込み)
for (i in 1:9) {
    cat(sprintf("7 × %d = %d\n", i, 7*i))
}
```

## if文:条件で分ける

```r
n <- 10
if (n %% 2 == 0) {
    cat(n, "は偶数\n")
} else {
    cat(n, "は奇数\n")
}
```

## for文 + if文 = 「探す」道具

さらにRらしい方法:**ベクトル演算**なら1行で「探せ」ます!

```r
# 1〜1000の中で7の倍数を全部表示
for (n in 1:1000) {
    if (n %% 7 == 0) print(n)
}
```

```r
# Rらしい書き方:条件で絞り込み
v <- 1:1000
v[v %% 7 == 0]
```

```r
# 個数と合計も一発!
c(sum(v %% 7 == 0), sum(v[v %% 7 == 0]))
```
