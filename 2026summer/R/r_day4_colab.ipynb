## グラフを描く準備

Rは**グラフ機能が標準装備**! パッケージのインストールは不要です。
(日本語ラベルは文字化けすることがあるため英語ラベルを使います)

## 関数の定義

```r
f <- function(x) 2*x + 1
g <- function(x) x^2 - 3*x + 2

c(f(3), g(1))
```

## グラフを描いてみよう!

`curve` 関数が便利です。

```r
curve(f, -3, 3, lwd = 3, col = "blue", main = "f(x) = 2x + 1")
grid()
```

```r
# 2つのグラフを重ねる(add = TRUE)
curve(f, -1, 6, lwd = 3, col = "blue")
curve(g, -1, 6, lwd = 3, col = "red", add = TRUE)
abline(h = 0)
grid()
```

## パラメータを動かして実験

```r
curve(x^2 - 4*x, -5, 5, lwd = 2, col = 1, ylim = c(-10, 30))
for (b in c(-2, 0, 2, 4)) {
    curve(x^2 + b*x, -5, 5, lwd = 2, col = b/2 + 4, add = TRUE)
}
grid()
```

## 最大・最小を求めよう

### 問題1:$f(x) = -x^2 + 4x + 1$ の最大値

```r
f1 <- function(x) -x^2 + 4*x + 1
curve(f1, -1, 5, lwd = 3, col = "blue")
grid()
```

```r
# 数値実験
xs <- seq(-1, 5, by = 0.001)
ys <- f1(xs)
cat("最大値:", max(ys), "(x =", xs[which.max(ys)], ")\n")
```

### 問題2:定義域つき $f(x) = x^2 - 4x + 3$(0 ≤ x ≤ 3)

```r
f2 <- function(x) x^2 - 4*x + 3
xs <- seq(0, 3, by = 0.001)
ys <- f2(xs)
plot(xs, ys, type = "l", lwd = 3, col = "purple")
grid()
cat("最大値:", max(ys), "(x =", xs[which.max(ys)], ")\n")
cat("最小値:", min(ys), "(x =", xs[which.min(ys)], ")\n")
```

### 問題3:3次関数 $f(x) = x^3 - 3x$(高2の先取り!)

```r
f3 <- function(x) x^3 - 3*x
curve(f3, -2.5, 2.5, lwd = 3, col = "darkgreen")
abline(h = 0)
grid()
```

## 応用:文章題を「描いて」解く

長さ20mのロープで長方形の囲い。面積 $S(x) = x(10-x)$ の最大は?

```r
S <- function(x) x * (10 - x)
curve(S, 0, 10, lwd = 3, col = "orange", xlab = "tate (m)", ylab = "menseki (m2)")
grid()
xs <- seq(0, 10, by = 0.001)
cat("最大面積:", max(S(xs)), "m2 (縦 =", xs[which.max(S(xs))], "m)\n")
```
