## コイン投げのシミュレーション

Rは**統計の言語**。乱数とシミュレーションは大得意です!

```r
runif(1)  # 0以上1未満の乱数(実行のたびに変わる!)
```

```r
# 10回のコイン投げ
for (i in 1:10) {
    cat(i, "回目:", if (runif(1) < 0.5) "表" else "裏", "\n")
}
```

```r
# 回数を増やすと0.5に近づく?(meanで一発!)
for (n in c(10, 100, 1000, 10000, 100000, 1000000)) {
    cat(n, "回 → 表の割合", mean(runif(n) < 0.5), "\n")
}
```

```r
# 大数の法則をグラフで
n <- 10000
results <- runif(n) < 0.5
ratios <- cumsum(results) / (1:n)
plot(ratios, type = "l", ylim = c(0.3, 0.7), xlab = "n", ylab = "ratio")
abline(h = 0.5, col = "red", lwd = 2)
```

## サイコロのシミュレーション

`sample(1:6, n, replace = TRUE)` でサイコロn回。`table` で集計も一発!

```r
n <- 60000
rolls <- sample(1:6, n, replace = TRUE)
table(rolls)  # 各目の回数(理論値は n/6 = 10000)
```

```r
barplot(table(rolls) / n, col = "skyblue", main = "dice")
abline(h = 1/6, col = "red", lwd = 2)
```

```r
# 2つのサイコロの和
n <- 100000
sums <- sample(1:6, n, replace = TRUE) + sample(1:6, n, replace = TRUE)
barplot(table(sums) / n, col = "orange", main = "sum of two dice")
```

## 誕生日問題:直感 vs 確率

```r
# 40人で同じ誕生日のペアがいる確率(replicateで10000回実験)
kekka <- replicate(10000, any(duplicated(sample(1:365, 40, replace = TRUE))))
cat("実験値:", mean(kekka), "\n")

# 理論値
p <- 1
for (i in 0:39) p <- p * (365 - i) / 365
cat("理論値:", 1 - p, "\n")
```

```r
# 人数を変えて理論値をグラフに
ks <- 1:60
riron <- sapply(ks, function(k) 1 - prod((365 - 0:(k-1)) / 365))
plot(ks, riron, type = "l", lwd = 2, xlab = "people", ylab = "probability")
abline(h = 0.5, col = "gray")
abline(v = 23, col = "red", lty = 2)  # 23人で50%超え!
```

## モンティ・ホール問題

```r
# 安全に1つ選ぶ関数(sample(x)はxが長さ1のとき注意が必要)
pick1 <- function(v) v[sample(length(v), 1)]

monty_hall <- function(change) {
    atari <- pick1(1:3)
    choice <- pick1(1:3)
    host <- pick1(setdiff(1:3, c(atari, choice)))
    if (change) choice <- setdiff(1:3, c(choice, host))
    choice == atari
}

n <- 100000
cat("変えない戦略の勝率:", mean(replicate(n, monty_hall(FALSE))), "\n")
cat("変える戦略の勝率:  ", mean(replicate(n, monty_hall(TRUE))), "\n")
```

## モンテカルロ法:乱数で円周率

```r
# 点を2000個打って可視化
n <- 2000
xs <- runif(n); ys <- runif(n)
inside <- xs^2 + ys^2 <= 1
plot(xs[inside], ys[inside], col = "red", pch = 20, cex = 0.5,
     xlim = c(0,1), ylim = c(0,1), asp = 1, xlab = "", ylab = "")
points(xs[!inside], ys[!inside], col = "blue", pch = 20, cex = 0.5)
```

```r
monte_carlo_pi <- function(n) 4 * mean(runif(n)^2 + runif(n)^2 <= 1)

for (n in c(100, 10000, 1000000, 10000000)) {
    cat(n, "個の点 → π ≈", monte_carlo_pi(n), "\n")
}
cat("本当のπ  =", pi, "\n")
```
