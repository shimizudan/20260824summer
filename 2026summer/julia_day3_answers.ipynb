## 問題1: 組合せの実験

6人(A〜F)から2人選ぶ方法を全部表示し、個数が $_6\mathrm{C}_2$ と一致することを確かめよう。

**解説**:全部で **15通り**。$_6\mathrm{C}_2 = \dfrac{6 \times 5}{2 \times 1} = 15$ と一致します。

```julia
# 6人から2人選ぶ --- 全部表示
people = ["A", "B", "C", "D", "E", "F"]
for c in combinations(people, 2)
    println(join(c))
end
```

```julia
# 個数の確認
length(collect(combinations(people, 2))), binomial(6, 2)
```

## 問題2: 同じものを含む順列

「JOHOKU」の6文字を並べ替える方法は何通り?(Oが2個)

**解説**:6文字のうちOが2個なので

$$\frac{6!}{2!} = \frac{720}{2} = 360 \text{ 通り}$$

```julia
# 1. 全部作って数える(collectで文字のリストにする)
moji = collect("JOHOKU")
length(unique(permutations(moji)))
```

```julia
# 2. 公式と比較
factorial(6) ÷ factorial(2)
```

```julia
# 参考:最初の10通りだけ表示してみる
for p in Iterators.take(unique(permutations(moji)), 10)
    println(join(p))
end
```

## 問題3: サイコロ2個の実験

大小2個のサイコロで目の和が7になる場合を全部表示し、何通りか数えよう。

**解説**:(1,6), (2,5), (3,4), (4,3), (5,2), (6,1) の **6通り**。
全36通り中6通りなので、確率は $\frac{6}{36} = \frac{1}{6}$(Day 5 につながります!)

```julia
# 和が7になる場合を全部表示
cnt = 0
for a in 1:6, b in 1:6
    if a + b == 7
        println("($a, $b)")
        cnt += 1
    end
end
println("全部で $cnt 通り")
```

## 問題4: チャレンジ問題

パスカルの三角形の各行の奇数の個数を数えてみよう(n = 0〜16)。

**解説**:結果は 1, 2, 2, 4, 2, 4, 4, 8, 2, 4, 4, 8, 4, 8, 8, 16, 2 —— **すべて2の累乗**!

実は「$n$ を2進法で表したときの1の個数を $k$ とすると、奇数の個数は $2^k$」という美しい法則があります(クンマーの定理)。
例:$n=7=111_{(2)}$ → $2^3=8$ 個、$n=16=10000_{(2)}$ → $2^1=2$ 個。

```julia
# 各行の奇数の個数
for n in 0:16
    kosuu = count(isodd, [binomial(n, r) for r in 0:n])
    println("n=$n: $kosuu 個")
end
```

```julia
# 発展:2進法の1の個数との関係を確認
for n in 0:16
    kosuu = count(isodd, [binomial(n, r) for r in 0:n])
    k = count(==(1), digits(n, base=2))
    println("n=$n: 奇数 $kosuu 個, 2進法 $(string(n, base=2)) の1は $k 個 → 2^$k = $(2^k)")
end
```
