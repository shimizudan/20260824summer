## 計算チェック!

展開の公式:$(2\sqrt{3}+5)(\sqrt{3}-1) = 1 + 3\sqrt{3}$

```python
# 左辺と右辺
(2*math.sqrt(3) + 5) * (math.sqrt(3) - 1), 1 + 3*math.sqrt(3)
```

```python
# ほぼ等しいかチェック(Juliaの≈に相当)
math.isclose((2*math.sqrt(3) + 5) * (math.sqrt(3) - 1), 1 + 3*math.sqrt(3))
```

## range:数列を一気に扱う

`range(1, 11)` は「1から10まで」(**終わりの数は含まない**ことに注意!)

```python
list(range(1, 11))  # 1〜10
```

```python
sum(range(1, 101))  # 1+2+...+100
```

```python
math.prod(range(1, 11))  # 10!
```

```python
list(range(1, 20, 2))  # 奇数:1から2ずつ
```

## for文:繰り返しの魔法

Pythonはブロックを**インデント(字下げ)**で表します(Juliaの end は不要)。

```python
for i in range(1, 6):
    print(i)
```

```python
# 7の段の九九(f文字列で埋め込み)
for i in range(1, 10):
    print(f"7 × {i} = {7*i}")
```

## if文:条件で分ける

```python
n = 10
if n % 2 == 0:
    print(f"{n} は偶数")
else:
    print(f"{n} は奇数")
```

```python
n = 100
if n % 3 == 0:
    print(f"{n} は3の倍数")
elif n % 3 == 1:
    print(f"{n} は3で割ると1余る")
else:
    print(f"{n} は3で割ると2余る")
```

## for文 + if文 = 「探す」道具

```python
# 1〜1000の中で7の倍数を全部表示!
for n in range(1, 1001):
    if n % 7 == 0:
        print(n)
```

```python
# 個数と合計
len([n for n in range(1, 1001) if n % 7 == 0]), sum(n for n in range(1, 1001) if n % 7 == 0)
```
