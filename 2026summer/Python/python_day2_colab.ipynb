## 約数を全部求めよう

```python
def divisors(n):
    return [d for d in range(1, n + 1) if n % d == 0]

divisors(24)
```

```python
len(divisors(24)), sum(divisors(24))  # 約数の個数と和
```

## 素数判定と sympy

Colabには数学パッケージ **sympy** が最初から入っています。

```python
# 自作:約数がちょうど2個なら素数
def isprime_mine(n):
    return n >= 2 and len(divisors(n)) == 2

isprime_mine(17), isprime_mine(21)
```

```python
from sympy import isprime, primerange, factorint

isprime(9999999967)
```

```python
list(primerange(1, 100))  # 100以下の素数
```

```python
factorint(2026)  # 素因数分解 {素数: 指数}
```

## エラトステネスの篩(ふるい)

```python
def eratosthenes(n):
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False
    for p in range(2, n + 1):
        if is_prime[p]:
            for q in range(2*p, n + 1, p):
                is_prime[q] = False
    return [p for p in range(n + 1) if is_prime[p]]

eratosthenes(100)
```

```python
# sympyの結果と一致するか確認
eratosthenes(1000) == list(primerange(1, 1001))
```

```python
# 素数の割合の実験
for n in [100, 1000, 10000, 100000]:
    k = len(list(primerange(1, n + 1)))
    print(f"{n} 以下の素数: {k} 個(割合 {k/n*100:.2f}%)")
```

## 最大公約数とユークリッドの互除法

```python
import math
math.gcd(1071, 1029)  # 標準装備
```

```python
def my_gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

my_gcd(1071, 1029)
```

```python
import math
math.lcm(12, 18), math.gcd(12, 18) * math.lcm(12, 18) == 12 * 18
```

## 完全数を探そう

```python
def proper_sum(n):
    return sum(divisors(n)) - n

for n in range(2, 10001):
    if proper_sum(n) == n:
        print(f"{n} は完全数! 約数: {divisors(n)}")
```

## 双子素数を探そう

```python
for n in range(2, 101):
    if isprime(n) and isprime(n + 2):
        print(f"({n}, {n+2})")
```

## フェルマー数

Pythonの整数は**最初から何桁でもOK**(Juliaの big() は不要)。

```python
for n in range(6):
    F = 2**(2**n) + 1
    print(f"F_{n} = {F} → 素数? {isprime(F)}")
```

```python
factorint(2**32 + 1)  # F₅の素因数分解(オイラーの発見を1秒で!)
```

## ピタゴラス数を探そう

```python
for c in range(1, 51):
    for b in range(1, c + 1):
        for a in range(1, b + 1):
            if a*a + b*b == c*c:
                print(f"({a}, {b}, {c})")
```

## コラッツ予想:世界一有名な未解決問題

```python
def collatz_steps(n):
    steps = 0
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3*n + 1
        steps += 1
    return steps

# 1〜30のステップ数
for n in range(1, 31):
    print(f"{n} → {collatz_steps(n)} ステップ")
```

```python
# 1〜10000で一番長い旅をする数は?
n_max = max(range(1, 10001), key=collatz_steps)
print(f"{n_max} が最長で {collatz_steps(n_max)} ステップ!")
```
