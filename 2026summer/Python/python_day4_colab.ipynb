## 関数の定義

```python
def f(x):
    return 2*x + 1

def g(x):
    return x**2 - 3*x + 2

f(3), g(1)
```

## グラフを描いてみよう!

```python
xs = np.linspace(-3, 3, 400)
plt.plot(xs, f(xs), lw=3, label="f(x) = 2x + 1")
plt.legend(); plt.grid(); plt.show()
```

```python
# 2つのグラフを重ねる
xs = np.linspace(-1, 6, 400)
plt.plot(xs, f(xs), lw=3, label="f(x)")
plt.plot(xs, g(xs), lw=3, label="g(x)")
plt.axhline(0, color="black", lw=1)
plt.legend(); plt.grid(); plt.show()
```

## パラメータを動かして実験

```python
xs = np.linspace(-5, 5, 400)
for b in range(-4, 5, 2):
    plt.plot(xs, xs**2 + b*xs, label=f"b = {b}")
plt.legend(); plt.grid(); plt.show()
```

## 最大・最小を求めよう

### 問題1:$f(x) = -x^2 + 4x + 1$ の最大値

```python
def f1(x):
    return -x**2 + 4*x + 1

xs = np.linspace(-1, 5, 400)
plt.plot(xs, f1(xs), lw=3, color="blue")
plt.grid(); plt.show()
```

```python
# 数値実験
xs = np.arange(-1, 5, 0.001)
ys = f1(xs)
print("最大値:", ys.max(), "(x =", xs[ys.argmax()], ")")
```

### 問題2:定義域つき $f(x) = x^2 - 4x + 3$(0 ≤ x ≤ 3)

```python
def f2(x):
    return x**2 - 4*x + 3

xs = np.arange(0, 3.001, 0.001)
ys = f2(xs)
plt.plot(xs, ys, lw=3, color="purple"); plt.grid(); plt.show()
print("最大値:", ys.max(), "(x =", xs[ys.argmax()], ")")
print("最小値:", ys.min(), "(x =", xs[ys.argmin()], ")")
```

### 問題3:3次関数 $f(x) = x^3 - 3x$(高2の先取り!)

```python
def f3(x):
    return x**3 - 3*x

xs = np.linspace(-2.5, 2.5, 400)
plt.plot(xs, f3(xs), lw=3, color="green")
plt.axhline(0, color="black", lw=1)
plt.grid(); plt.show()
```

## 応用:文章題を「描いて」解く

長さ20mのロープで長方形の囲い。面積 $S(x) = x(10-x)$ の最大は?

```python
def S(x):
    return x * (10 - x)

xs = np.arange(0, 10.001, 0.001)
plt.plot(xs, S(xs), lw=3, color="orange")
plt.xlabel("縦の長さ (m)"); plt.ylabel("面積 (m²)")
plt.grid(); plt.show()
print("最大面積:", S(xs).max(), "m² (縦 =", xs[S(xs).argmax()], "m)")
```
