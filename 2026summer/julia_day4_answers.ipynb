## 問題1: グラフを描こう

$f(x) = x^2 - 6x + 7$ のグラフを描き、頂点の座標を読み取ろう。

**解説**:平方完成すると $f(x) = (x-3)^2 - 2$ なので、頂点は $(3, -2)$、最小値は $-2$。

```julia
# グラフを描く
f(x) = x^2 - 6x + 7
plot(f, -1, 7, lw=3, color=:blue, label="f(x) = x² - 6x + 7")
```

```julia
# 数値実験で最小値
xs = -1:0.001:7
ys = f.(xs)
println("最小値: $(minimum(ys)) (x = $(xs[argmin(ys)]))")
```

## 問題2: 定義域つきの最大・最小

$f(x) = -x^2 + 2x + 3$($0 \le x \le 4$)の最大値と最小値。

**解説**:$f(x) = -(x-1)^2 + 4$。頂点 $x=1$ で**最大値4**、端 $x=4$ で**最小値** $f(4) = -5$。

```julia
# グラフで確認
g(x) = -x^2 + 2x + 3
plot(g, 0, 4, lw=3, color=:purple, label="g(x) = -x² + 2x + 3 (0 ≤ x ≤ 4)")
```

```julia
# 数値実験
xs = 0:0.001:4
ys = g.(xs)
println("最大値: $(maximum(ys)) (x = $(xs[argmax(ys)]))")
println("最小値: $(minimum(ys)) (x = $(xs[argmin(ys)]))")
```

## 問題3: 文章題

長さ40cmの針金を折り曲げて長方形を作る。面積が最大になるときの縦の長さと最大面積。

**解説**:縦を $x$ とすると横は $20 - x$。面積 $S(x) = x(20-x) = -(x-10)^2 + 100$。
縦 **10cm**(正方形)のとき最大面積 **100cm²**。

```julia
# 面積の関数を定義して描く
S(x) = x * (20 - x)
plot(S, 0, 20, lw=3, color=:orange, label="S(x) = x(20-x)", xlabel="縦の長さ (cm)", ylabel="面積 (cm²)")
```

```julia
# 数値実験で最大値
xs = 0:0.001:20
println("最大面積: $(maximum(S.(xs))) cm² (縦 = $(xs[argmax(S.(xs))]) cm)")
```

## 問題4: チャレンジ問題

$f(x) = x + \dfrac{4}{x}$($x > 0$)の最小値。

**解説**:相加平均・相乗平均の関係から

$$x + \frac{4}{x} \ge 2\sqrt{x \cdot \frac{4}{x}} = 4$$

等号は $x = \frac{4}{x}$、つまり $x = 2$ のとき。**最小値4**が実験でも確認できます。

```julia
# グラフを描く
h(x) = x + 4/x
plot(h, 0.5, 8, lw=3, color=:green, label="h(x) = x + 4/x")
```

```julia
# 数値実験で最小値
xs = 0.1:0.0001:10
ys = h.(xs)
println("最小値: $(minimum(ys)) (x = $(xs[argmin(ys)]))")
```
