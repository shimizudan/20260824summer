## 約数を全部求めよう

$n$ の約数とは、$n$ を割り切る正の整数のこと。「$n \% d == 0$ となる $d$ を探す」だけです。

```julia
# 24の約数を全部表示
for d in 1:24
    if 24 % d == 0
        println(d)
    end
end
```

```julia
# 関数にしてしまおう! 約数のリストを返す
divisors(n) = [d for d in 1:n if n % d == 0]

divisors(24)
```

```julia
# 約数の個数
length(divisors(24))
```

```julia
# 約数の和
sum(divisors(24))
```

```julia
# いろいろな数で実験してみよう
for n in [12, 36, 60, 100]
    println("$n の約数: $(divisors(n))")
end
```

## 素数判定

素数とは「1と自分自身以外に約数を持たない、2以上の整数」。つまり**約数がちょうど2個**の数です。

```julia
# 約数の個数が2個なら素数!
isprime_mine(n) = n ≥ 2 && length(divisors(n)) == 2

isprime_mine(17), isprime_mine(21)
```

```julia
# 1から30までの素数を表示
for n in 1:30
    if isprime_mine(n)
        println(n)
    end
end
```

## エラトステネスの篩（ふるい）

紀元前3世紀の数学者エラトステネスが考えた、素数を**ふるい落として**見つける方法です。

1. 2から $n$ までの数を並べる
2. 残っている最小の数（素数）を残し、その倍数を全部消す
3. 繰り返す

```julia
# エラトステネスの篩を実装してみよう
function eratosthenes(n)
    is_prime = trues(n)      # 全部「素数候補」でスタート
    is_prime[1] = false      # 1は素数ではない
    for p in 2:n
        if is_prime[p]
            for q in 2p:p:n  # pの倍数をふるい落とす
                is_prime[q] = false
            end
        end
    end
    return [p for p in 1:n if is_prime[p]]
end

eratosthenes(100)
```

```julia
# Primes.jlの結果と一致するか確認
eratosthenes(1000) == primes(1000)
```

```julia
# 素数はどのくらいの割合であるのか実験
for n in [100, 1000, 10000, 100000]
    println("$n 以下の素数: $(length(primes(n))) 個（割合 $(round(length(primes(n))/n*100, digits=2))%）")
end
```

## 最大公約数とユークリッドの互除法

紀元前300年頃の「原論」に載っている、世界最古のアルゴリズムの一つです。

**原理**：$a$ を $b$ で割った余りを $r$ とすると、$\gcd(a, b) = \gcd(b, r)$

```julia
# Juliaには gcd が標準装備
gcd(1071, 1029)
```

```julia
# ユークリッドの互除法を自分で書いてみよう
function my_gcd(a, b)
    while b != 0
        println("gcd($a, $b)")   # 途中経過を表示
        a, b = b, a % b
    end
    return a
end

my_gcd(1071, 1029)
```

```julia
# 最小公倍数も標準装備
lcm(12, 18)
```

```julia
# 確認：gcd × lcm = 2数の積
gcd(12,18) * lcm(12,18) == 12 * 18
```

## 完全数を探そう

**完全数**とは「自分以外の約数の和が自分自身と等しい数」。

例：$6 = 1 + 2 + 3$

古代ギリシャ時代から知られる神秘的な数です。次の完全数を**探して**みましょう。

```julia
# 自分以外の約数の和
proper_sum(n) = sum(divisors(n)) - n

proper_sum(6)
```

```julia
# 10000以下の完全数を全部探す!
for n in 2:10000
    if proper_sum(n) == n
        println("$n は完全数! 約数: $(divisors(n))")
    end
end
```

発見した完全数 6, 28, 496, 8128 … 実はすべて $2^{p-1}(2^p - 1)$ の形（$2^p-1$ が素数のとき）をしています。

奇数の完全数が存在するかどうかは、**現在も未解決**です!

## 双子素数を探そう

**双子素数**とは、差が2の素数のペア。例：(3,5), (5,7), (11,13) …

「双子素数は無限にあるか?」も**未解決問題**です。

```julia
# 100以下の双子素数を全部探す
for n in 2:100
    if isprime(n) && isprime(n+2)
        println("($n, $(n+2))")
    end
end
```

## フェルマー数

フェルマーは $F_n = 2^{2^n} + 1$ がすべて素数だと予想しました。

$F_0=3,\ F_1=5,\ F_2=17,\ F_3=257,\ F_4=65537$ … 確かに素数。では $F_5$ は?

```julia
# フェルマー数を計算して素数判定
for n in 0:5
    F = big(2)^(2^n) + 1   # bigで大きな整数を扱う
    println("F_$n = $F → 素数? $(isprime(F))")
end
```

```julia
# F_5 を素因数分解してみよう（フェルマーの予想は崩れた!）
factor(big(2)^(2^5) + 1)
```

オイラーが1732年に $F_5 = 641 × 6700417$ を発見し、フェルマーの予想は崩れました。

## ピタゴラス数を探そう

$a^2 + b^2 = c^2$ を満たす自然数の組 $(a, b, c)$ を**ピタゴラス数**といいます。例：$(3, 4, 5)$

```julia
# c ≤ 50 のピタゴラス数を全部探す（a < b とする）
for c in 1:50, b in 1:c, a in 1:b
    if a^2 + b^2 == c^2
        println("($a, $b, $c)")
    end
end
```

## コラッツ予想：世界一有名な未解決問題

どんな自然数 $n$ から始めても、

- 偶数なら **2で割る**
- 奇数なら **3倍して1を足す**

を繰り返すと、**必ず1にたどり着く**（と予想されている）。

1937年に提出されてから、**誰も証明できていません**。しかしJuliaなら数行で実験できます!

```julia
# コラッツの操作を繰り返して、1になるまでの道のりを表示
function collatz(n)
    steps = 0
    while n != 1
        n = iseven(n) ? n ÷ 2 : 3n + 1
        print(n, " ")
        steps += 1
    end
    println()
    return steps
end

collatz(6)
```

```julia
# 27から始めるとかなり長い旅に!
collatz(27)
```

```julia
# 1になるまでのステップ数だけ返すバージョン
function collatz_steps(n)
    steps = 0
    while n != 1
        n = iseven(n) ? n ÷ 2 : 3n + 1
        steps += 1
    end
    return steps
end

# 1〜30のステップ数を実験
for n in 1:30
    println("$n → $(collatz_steps(n)) ステップ")
end
```

```julia
# 1〜10000で一番長くかかる数はどれ?
n_max = argmax(collatz_steps, 1:10000)
println("$n_max が最長で $(collatz_steps(n_max)) ステップ!")
```

現在、$2^{68}$（約30京）までの数はすべて1に到達することが**コンピュータで**確認されています。
