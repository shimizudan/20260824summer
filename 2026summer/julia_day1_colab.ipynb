## Julia言語とは

## for文：繰り返しの魔法

同じ処理を何度も繰り返すには `for` 文を使います。

```julia
for 変数 in 範囲
    処理
end
```

````julia
# 1から5まで表示
for i in 1:5
    println(i)
end
````

````julia
# 7の段の九九
for i in 1:9
    println("7 × $i = $(7i)")
end
````

````julia
# 1から10までの2乗の値
for i in 1:10
    println("$i の2乗は $(i^2)")
end
````

## if文：条件で分ける

条件によって処理を変えるには `if` 文を使います。

```julia
if 条件
    処理A
else
    処理B
end
```

条件には `==`（等しい）、`<`、`>`、`%`（余り）などが使えます。

````julia
# 偶数か奇数か判定
n = 10
if n % 2 == 0
    println("$n は偶数")
else
    println("$n は奇数")
end
````

````julia
# 3で割った余りで分類
n = 100
if n % 3 == 0
    println("$n は3の倍数")
elseif n % 3 == 1
    println("$n は3で割ると1余る")
else
    println("$n は3で割ると2余る")
end
````

## for文 + if文 = 「探す」道具

この2つを組み合わせると、**条件に合う数を全部探す**ことができます。これが「実験する数学」の第一歩です!

````julia
# 1〜1000の中で7の倍数を全部表示してみよう!
for n in 1:1000
    if n % 7 == 0
        println(n)
    end
end
````

````julia
# 何個あるかは count で一発!
count(n -> n % 7 == 0, 1:1000)
````

````julia
# 7の倍数の合計も求められる
sum(n for n in 1:1000 if n % 7 == 0)
````
